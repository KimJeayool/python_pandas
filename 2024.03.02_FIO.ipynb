```python
arr = [80, 90, 87, 78, 86]

def average(scores):
    avg = sum(scores) / len(scores)
    return avg

def grade(avg):
    if avg >= 90:
        return 'A'
    if avg >= 80:
        return 'B'
    if avg >= 70:
        return 'C'
    if avg >= 60:
        return 'D'
    else:
        return 'F'
```

```python
# name 인덱스 축으로 지정
# 국어~역사
df = pd.read_csv('csv/exam1.csv', index_col=0)

df['avg'] = df.loc[:, 'kor':].apply(average, axis = 'columns')
df['grd'] = df['avg'].apply(grade)
df.to_csv('csv/result1.csv', encoding='utf-8-sig')
```
