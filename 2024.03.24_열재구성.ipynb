```python
# 삼성전자 주식 정보
stock = pd.read_csv('csv/삼성전자_주식.csv')
# 날짜형 -> 문자열
stock['date'] = pd.to_datetime(stock['date'])
stock.info()
```

```python
# 날짜형 -> 문자열
stock['date'] = stock['date'].astype(str)
print(stock.dtypes)

# 날짜
dates = stock['date'].str.split('-')

# 년월일
stock['year'] = dates.str.get(0)
stock['year'] = stock['year'].astype(int)
stock['month'] = dates.str.get(1)
stock['month'] = stock['month'].astype(int)
stock['day'] = dates.str.get(2)
stock['day'] = stock['day'].astype(int)

print(stock.dtypes)
```

```python
# 사용자 정의
def str_to_year(text):
    return int(text.split('-')[0])
def str_to_month(text):
    return int(text.split('-')[1])
def str_to_day(text):
    return int(text.split('-')[2])

def str_to_date(text):
    date = text.split('-')
    year = int(date[0])
    month = int(date[1])
    day = int(date[2])
    return pd.Series({'year': year, 'month': month, 'day':day})


stock = pd.read_csv('csv/삼성전자_주식.csv')
stock['연'] = stock['date'].apply(str_to_year)
stock['월'] = stock['date'].apply(str_to_month)
stock['일'] = stock['date'].apply(str_to_day)

stock = pd.concat([stock, stock['date'].apply(str_to_date)], axis='columns')

stock.dtypes
```
