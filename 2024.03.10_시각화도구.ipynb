```python
# 라이브러리 불러오기
import pandas as pd
import requests
import os
from bs4 import BeautifulSoup
from time import sleep

# 시각화
from matplotlib import font_manager
from matplotlib import rc
import matplotlib.pyplot as plt

# 경고 무시
import warnings
warnings.filterwarnings('ignore')

# 한글 인코딩
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 데이터 재정의
df = pd.read_excel('excel/시도별_전출입_인구수.xlsx')
df = df.drop(0, axis='index')

# 전출지별 데이터 체우기
df['전출지별'] = df['전출지별'].ffill()
df = df.rename({'전출지별':'전출지', '전입지별':'전입지'}, axis='columns')
df.head(3)
```

```python
# 인덱스 번호로 뽑는 방법 (비권장) 
df2 = df.iloc[18:35]
df2 = df2.drop(20, axis='index')
df2 = df2.drop('전출지', axis='columns')
df2 = df2.set_index('전입지')
df2
```

```python
# 조건문으로 뽑는 방법 (권장)
mask = (df['전출지'] == '서울특별시') & (df['전입지'] != '서울특별시')
df3 = df[mask]
df3 = df3.drop('전출지', axis='columns')
df3 = df3.set_index('전입지')

df3.head(3)
```

```python
# 서울 -> 경기도 전입
sr1 = df3.loc['경기도']
sr1 = sr1.map(int)
print(sr1[:5])

# 그래프
year = sr1.index
data = sr1.values
plt.plot(year, data)
```

```python
# X축 컬럼 숫자형으로 변경
sr1.index = sr1.index.map(int)

# 그래프 꾸미기
plt.figure(figsize=(10,5))
plt.title('서울시 -> 경기도 인구이동', size=18)

# 그래프 생성
plt.plot(sr1)
# 그래프 출력
plt.show()
```

```python
# 데이터 뽑기
sr2 = df3.loc['경기도']
sr2 = sr2.map(int)

# 2000년 ~ 2023년 
# 서울 -> 경기 인구 이동
sr2 = sr2['2000':]

# 그래프 크기
plt.figure(figsize=(8,3))

# 그래프 생성
plt.plot(sr2)

# 그래프 꾸미기
plt.title('서울특별시 -> 경기도 인구 이동 (2000~2023)', size=16)
plt.xticks(rotation=65, size=11)
plt.xlabel('기간', size=12)
plt.ylabel('인구수', size=12)
plt.legend(labels = ['서울->경기'], loc='best')

# 그래프 출력
plt.show()
```
