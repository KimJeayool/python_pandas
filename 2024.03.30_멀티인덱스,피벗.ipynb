```python
# 라이브러리 불러오기
import pandas as pd
import seaborn as sns
import warnings
from numpy import nan

# 경고 무시
import warnings
warnings.filterwarnings('ignore')
```

```python
# 타이타닉 승객 정보
titanic = sns.load_dataset('titanic')
titanic.head(2)
```

```python
# 데이터 추출
df = titanic.groupby(['class', 'sex']).agg({'age':'mean', 'fare':['max', 'min']})
df
```

```python
# 첫번째 클래스의 여성만 조회.
df.loc[('First', 'female')]
```

```python
# 각 클래스의 여성만 조회.
df.loc[[('First', 'female'), ('Second', 'female'), ('Third', 'female')]]
```

```python
# 인덱스 초기화
df2 = df.reset_index()
# 마스크 적용
mask = df2['sex'] == 'female'
df2 = df2[mask]
df2.set_index(['class', 'sex'])
```

```python
# 여러 인덱스를 바로 사용 가능.
df.xs('female', level='sex')
```

```python
# 삼성전자 주식 정보
stock = pd.read_csv('csv/삼성전자_주식.csv')
stock.info()
```

```python
# 데이터 복사
stock2 = stock.copy()

# 날짜 추출
sr_date = pd.to_datetime(stock['date'])
# 년월일
sr_year = sr_date.dt.year
sr_month = sr_date.dt.month
sr_day = sr_date.dt.day

# 삽입
stock2['year'] = sr_year
stock2['month'] = sr_month
stock2['day'] = sr_day

# 멀티 인덱스
stock = stock2.set_index(['year', 'month', 'day'])
stock
```

```python
# 1월5일~1월10일, 2월5일~2월10일, 3월5일~3월10일
dates = []
for month in range(1,4):
    for day in range(15, 16):
        dates.append((2024, month, day))

stock.loc[dates]
```

```python
# 컬럼 추출
columns = ['age', 'sex', 'class', 'fare', 'survived']

# 데이터 추출
df = titanic[columns]
df.head(3)
```

```python
'''
선실을 기준으로 성별별 평균 나이.
'''
pd.pivot_table(df, 
               index='class',
               columns='sex',
               values='age',
               aggfunc='mean')
```

```python
'''
선실 기준으로 성별별 평균 나이와 인원수
'''
pd.pivot_table(df,
               index='class',
               columns='sex',
               values='age',
               aggfunc=['mean', 'count'])
```

```python
'''
선실 기준으로 성별별 살아남은 비율과 인원수, 총 탑승객
'''
pd.pivot_table(df,
               index='class',
               columns='sex',
               values='survived',
               aggfunc=['mean', 'sum', 'count'])
```

```python
# 사용자 정의 합수
def mean2(sr):
    return sum(sr) / len(sr)

'''
행 기준 : 선실, 성별
열 기준 : 생존
데이터 : 나이, 요금
계산 : 평균, 최대
'''
df_pivot = pd.pivot_table(df,
               index=['class', 'sex'],
               columns='survived',
               values=['age', 'fare'],
               # aggfunc=[mean2, 'max'])
               aggfunc=['mean', 'max'])
df_pivot
```

```python
# 엑셀 파일 저장
df_pivot.to_excel('excel/titanic_pivot_table.xlsx')
```
