```python
# 라이브러리 불러오기
import pandas as pd
import seaborn as sns
import warnings
from numpy import nan
```

```python
# 타이타닉 정보
titanic = sns.load_dataset('titanic')
titanic.head(4)
```

```python
# 데이터 제한하여 추출
df = titanic.iloc[:5][['age', 'fare']]
df
```

```python
# 함수 - 더하기 10
def add_10(n):
    return n + 10

# 함수 적용
df['age'].apply(add_10)
```

```python
# 함수 정의 - 등급 반환
def grade(score):
    if score >= 90: grd='A'
    elif score >=80: grd='B'
    elif score >=70: grd='C'
    elif score >=60: grd='D'
    else: grd='F'
    return grd

# 학생 데이터
students = pd.read_csv('csv/class_1.csv', encoding='euckr', 
                       header=None,
                       names=['name','addr','kor','eng','math','sci','his'],
                       index_col='name')

# 평균
students['avg'] = students.loc[:, 'kor':].mean(axis='columns')
students['grade'] = students['avg'].apply(grade)
students
```

```python
# 함수 정의2
def add_two(x, y=1):
    return x + y
df['age'].apply(add_two, y=10)

# 함수 정의3
def add_three(x, y=0, z=0):
    return x  + y +z
df['age'].apply(add_three, y=10, z=5)

# 람다식
df['age'].apply(lambda x: x+10)
```

```python
# 함수 정의
def compute_sum(sr):
    return sr.sum()

# 함수 적용
df.apply(compute_sum)
```

```python
# 원소별 데이터 반환
students.loc[:, 'kor':'his'].applymap(grade)
```

```python
# 함수 정의 (시리즈 -> 시리즈)
def missing_values(sr):
    return sr.isnull()
    
# 함수 정의 (시리즈 -> 값)
def missing_values_count(sr):
    return sr.isnull().sum()

'''
누락 개수
df.isnull().sum()
len(df.index) - df.count()
'''
    
# 데이터 추출
df = titanic.iloc[15:30][['age', 'fare']]

# 시리즈 -> 시리즈
df.apply(missing_values, axis='index')
```

```python
# 시리즈 -> 값 (열 단위)
df.apply(missing_values_count, axis='index')
```

```python
# 시리즈 -> 값 (행 단위)
df.apply(missing_values_count, axis='columns')
```

```python
# 
students = pd.read_csv('csv/class_1.csv', encoding='euckr', header=None,
                       names=['name', 'addr', 'kor', 'eng', 'math', 'sci', 'his'],
                       index_col='name')
students
```

```python
# 함수 정의 : 평균 (시리즈 -> 값)
def average(sr):
    return sr.sum() / len(sr.index)

# 원본 복사
students2 = students.copy()

# 평균 컬럼 추가
students2['avg'] = students2.loc[:, 'kor':].apply(average, axis='columns')
students2
```

```python
# 함수 정의
def add_two_obj(sr1, sr2):
    return sr1 + sr2

# 원본 복사
df2 = df.copy()
df2['ten'] = 10

# 행 단위 시리즈 -> 시리즈 반환
df2['age+ten'] = df2.apply(lambda sr: sr['age'] + sr['ten'], axis='columns')
df2
```

```python
# 함수 정의
def sub3_average(sr):
    return (sr['kor'] + sr['eng'] + sr['math']) / 3

# 원본 복사
students2 = students.copy()

# 행 단위
print(students2.apply(sub3_average, axis='columns'))
print(students2.apply(lambda sr: (sr['kor'] + sr['eng'] + sr['math']) / 3, axis='columns'))
```

```python
# 함수 정의 (DF 반환)
def missing_value(df):
    return df.isnull()

# 함수 정의 (SR 반환)
def missing_count(df):
    return df.isnull().sum()

# 함수 정의 (값 반환)
def missing_total_count(df):
    return df.isnull().sum().sum()


# 데이터 추출
df = titanic.loc[15:24, ['age', 'fare', 'deck']]

# DF 반환
print(df.pipe(missing_value))
print('='*30)
# SR 반환
print(df.pipe(missing_count))
print('='*30)
# 값 반환
print(df.pipe(missing_total_count))
```
