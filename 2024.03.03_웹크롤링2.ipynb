```python
# 모듈 Import
import pandas as pd
import requests
from bs4 import BeautifulSoup
from time import sleep
```

```python
# 웹 문서 불러오기
# url = 'http://192.168.50.54/index.html'
url = 'https://namu.wiki/'
response = requests.get(url)

# 결과 출력
print(response)
print('===================')
print(response.headers)
print('===================')
print(response.encoding)
print('===================')
print(response.content)
print('===================')
print(response.text)
```

```python
# 웹 문서 내용
text = response.text
html = BeautifulSoup(text, 'html.parser')

# tr 목록
tr_list = html.find('tbody').find_all('tr')
# tr 데이터
rows = []
for tr in tr_list:
    # td 목록
    td_list = tr.find_all('td')
    row = []
    for td in td_list:
        # td 데이터
        row.append(td.text)
    # 결과 배열    
    rows.append(row)
# 결과 조회
print(rows)
```

```python
# 나무위키
url = 'http://192.168.50.54/professors.html'
# 요청
response = requests.get(url)
# 응답 확인
print(response.headers)

# 텍스트 변환
text = response.text
print(text)
```

```python
# 함수 선언
def make_row(tr):
    no = tr.find('td', {'class':'number'}).text
    name = tr.find('td', {'class':'professor'}).text
    lecture = tr.find('td', {'class':'lecture'}).text
    grade = tr.find('td', {'class':'grade'}).text
    evl = tr.find('td', {'class':'evaluation'}).text
    row = [no, name, lecture, grade, evl]
    return row

# 분석 객체
html = BeautifulSoup(text, 'html.parser')

# tr 목록 추출
tr_list = html.find('tbody').find_all('tr')

# tr 목록 조회
rows = []
# 데이터 수집
for tr in tr_list:
    row = make_row(tr)
    rows.append(row)

# 데이터프레임 생성
df = pd.DataFrame(rows, columns=['no', 'name', 'lecture', 'grade', 'evaluation'])
df.set_index('no', inplace=True)

# csv 파일 생성
df.to_csv('csv/professors.csv')

df
```
