```python
# Pands
import pandas as pd
```

```python
exam_data = {'수학':[ 90, 80,  70, 60,  70],
             '영어':[ 98, 89,  99, 90, 100],
             '음악':[ 85, 95, 100, 90,  90],
             '체육':[100, 90,  90, 80,  70]}

df = pd.DataFrame(exam_data, index = ['서준', '우현', '인아', '진수', '진희'])
df
```

```python
exam_data2 = {'이름':['서준', '우현', '인아', '진수', '진희'],
              '수학':[ 90, 80,  70, 60,  70],
              '영어':[ 98, 89,  99, 90, 100],
              '음악':[ 85, 95, 100, 90,  90],
              '체육':[100, 90,  90, 80,  70]}

df2 = pd.DataFrame(exam_data2)

# 원본 변경
df2.set_index('이름', inplace=True)
df2
```

```python
# 열(Columns) -> 행(Row)     : df['컬러명']['인덱스명']
# 행(Row)     -> 열(Columns) : df.loc['인덱스명', '컬러명']
#                           : df.loc['인덱스명']['컬러명'] 
print(df['영어']['인아'])
print(df.영어['인아'])
print(df.loc['진수']['영어'])
print(df.loc['진수', '영어'])
print(df.iloc[3, 1])
```
