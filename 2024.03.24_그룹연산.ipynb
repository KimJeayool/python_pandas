```python
# 라이브러리 불러오기
import pandas as pd
import seaborn as sns
import warnings
from numpy import nan

# 경고 무시
import warnings
warnings.filterwarnings('ignore')
```

```python
# 타이타닉 승객 정보
titanic = sns.load_dataset('titanic')
titanic.head(2)
```

```python
# 타이타닉 승객 정보
df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]
df.head(5)
```

```python
# 그룹
list(df.groupby('sex'))[0][1]

# 반복문
for key, group in df.groupby('class'):
    print('key: ', key)
    print('count: ', len(group))
    print(group.head(5))
```

```python
# 평균
grouped = df.groupby('class')
grouped.mean(numeric_only=True)
```

```python
# 2개 이상 기준
grouped2 = df.groupby(['class', 'sex'])

# 데이터 조회
for key, group in grouped2:
    print('key: ', key)
    print('count: ', len(group))
    print(group.head(5))
```

```python
# 그룹 목록
grouped = df.groupby(['class'])

# 표준 편차
grouped.std(numeric_only=True)
```

```python
# 특정 컬럼의 표준편차
grouped[['age', 'fare']].agg(['mean', 'std'])
```

```python
# 사용자 정의 함수
def min_max(x):
    return x.max() - x.min()

# 함수 적용
print(grouped[['age', 'fare']].agg(min_max))
print('='*100)
# 여러 함수 적용
print(grouped[['age', 'fare']].agg([max,min,min_max]))
print('='*100)
# 각 컬럼별로 함수 적용
print(grouped[['age', 'fare']].agg({'age':'mean', 'fare':'std'}))
print('='*100)
print(grouped[['age', 'fare']].agg({'age':['mean', 'std'], 'fare':['mean', 'max']}))
```

```python
# 데이터 추출
grouped = df.groupby('class')

# 데이터 조회
for key, group in grouped:
    print(key, len(group))
print('='*100)

# 필터 조건
grouped_filter = grouped.filter(lambda x: len(x) >= 200)
print(grouped_filter)
```

```python
# 사용자 정의 함수
def over_200(x):
    return len(x) > 200

# 필터 적용
print(grouped.filter(over_200))
```

```python
# 개별로 함수 적용
grouped.apply(lambda x: x.describe())
```

```python
# 나이 평균 30 미만 데이터 조회
grouped.filter(lambda df: df['age'].mean() < 30)
```

```python
# 나이 평균 30 미만 조건 생성
age_filter = grouped.apply(lambda x: x['age'].mean() < 30)

df_list = []
for x in age_filter.index:
    if age_filter[x] == True:
        age_filter_df = grouped.get_group(x)
        df_list.append(age_filter_df)
        
pd.concat(df_list)
```

```python
# 데이터 복사
titanic2 = titanic.copy()

# who 그룹 별 age 평균 나이
who_mean_age = titanic2.groupby('who')['age'].mean()
print(who_mean_age)
print('='*100)

# 성별로 NaN 평균나이 치환
df_list = []
for key in who_mean_age.index:
    mask = titanic2['who'] == key
    df = titanic2[mask]
    df['age'] = df['age'].fillna(who_mean_age[key])
    df_list.append(df)
dfs = pd.concat(df_list)
dfs.sort_index(inplace=True)

# 결과물 비교
ori_age = titanic2['age']
dfs.insert(4, 'ori_age', ori_age)
dfs[25:50]
```
